```python
from __future__ import annotations

import json
import math
import os
import random
import warnings
from dataclasses import dataclass
from pathlib import Path
from typing import Dict, List, Sequence, Tuple

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas.plotting import scatter_matrix
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    precision_recall_fscore_support,
)
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler

warnings.filterwarnings("ignore", category=UserWarning)

# ============================================================
# PROJECT SETTINGS
# ============================================================

PROJECT_DIR = Path(
    r"C:\Users\NXTWAVE\Downloads\classification of Iris flower"
)
OUTPUT_DIR = PROJECT_DIR / "outputs"

RANDOM_SEED = 42
BASELINE_EPOCHS = 1000
BATCH_SIZE = 16
BASELINE_LEARNING_RATE = 0.01
TRACK_EVERY = 50

np.random.seed(RANDOM_SEED)
random.seed(RANDOM_SEED)


# ============================================================
# GENERAL UTILITIES
# ============================================================

def create_output_folder() -> None:
    """Create the required project and output folders."""
    PROJECT_DIR.mkdir(parents=True, exist_ok=True)
    OUTPUT_DIR.mkdir(parents=True, exist_ok=True)


def one_hot_encode(labels: np.ndarray, number_of_classes: int) -> np.ndarray:
    """Convert integer class labels into one-hot encoded rows."""
    encoded = np.zeros((labels.shape[0], number_of_classes), dtype=float)
    encoded[np.arange(labels.shape[0]), labels] = 1.0
    return encoded


def save_text(path: Path, content: str) -> None:
    """Save UTF-8 text to disk."""
    path.write_text(content, encoding="utf-8")


# ============================================================
# DATA PREPARATION AND EDA
# ============================================================

def load_and_prepare_data():
    """
    Load Iris, create a DataFrame, split 70/15/15, standardize features,
    and one-hot encode labels.
    """
    iris = load_iris()

    feature_names = [
        "sepal_length_cm",
        "sepal_width_cm",
        "petal_length_cm",
        "petal_width_cm",
    ]

    class_names = list(iris.target_names)

    dataframe = pd.DataFrame(iris.data, columns=feature_names)
    dataframe["species"] = [class_names[index] for index in iris.target]

    x = iris.data.astype(float)
    y = iris.target.astype(int)

    # First split: 70% training and 30% temporary.
    x_train, x_temp, y_train, y_temp = train_test_split(
        x,
        y,
        test_size=0.30,
        random_state=RANDOM_SEED,
        stratify=y,
    )

    # Second split: temporary data becomes 15% validation and 15% test.
    x_validation, x_test, y_validation, y_test = train_test_split(
        x_temp,
        y_temp,
        test_size=0.50,
        random_state=RANDOM_SEED,
        stratify=y_temp,
    )

    scaler = StandardScaler()
    x_train_scaled = scaler.fit_transform(x_train)
    x_validation_scaled = scaler.transform(x_validation)
    x_test_scaled = scaler.transform(x_test)

    y_train_one_hot = one_hot_encode(y_train, len(class_names))
    y_validation_one_hot = one_hot_encode(y_validation, len(class_names))
    y_test_one_hot = one_hot_encode(y_test, len(class_names))

    return {
        "dataframe": dataframe,
        "feature_names": feature_names,
        "class_names": class_names,
        "x": x,
        "y": y,
        "x_train": x_train_scaled,
        "x_validation": x_validation_scaled,
        "x_test": x_test_scaled,
        "y_train": y_train,
        "y_validation": y_validation,
        "y_test": y_test,
        "y_train_one_hot": y_train_one_hot,
        "y_validation_one_hot": y_validation_one_hot,
        "y_test_one_hot": y_test_one_hot,
        "scaler": scaler,
    }


def save_eda_outputs(dataframe: pd.DataFrame, feature_names: Sequence[str]) -> None:
    """Save histograms, pair plot, correlation heatmap, and dataset summary."""

    # Feature histograms
    dataframe[list(feature_names)].hist(
        bins=15,
        figsize=(12, 8),
        edgecolor="black",
    )
    plt.suptitle("Iris Feature Distributions", fontsize=15)
    plt.tight_layout(rect=(0, 0, 1, 0.96))
    plt.savefig(OUTPUT_DIR / "feature_histograms.png", dpi=300, bbox_inches="tight")
    plt.close()

    # Pair plot using pandas scatter_matrix
    axes = scatter_matrix(
        dataframe[list(feature_names)],
        figsize=(11, 11),
        diagonal="hist",
        alpha=0.75,
    )
    plt.suptitle("Iris Feature Pair Plot", fontsize=15, y=0.92)
    plt.savefig(OUTPUT_DIR / "feature_pair_plot.png", dpi=300, bbox_inches="tight")
    plt.close()

    # Correlation heatmap
    correlation = dataframe[list(feature_names)].corr()
    fig, ax = plt.subplots(figsize=(8, 6))
    image = ax.imshow(correlation.values, cmap="coolwarm", vmin=-1, vmax=1)

    ax.set_xticks(range(len(feature_names)))
    ax.set_yticks(range(len(feature_names)))
    ax.set_xticklabels(feature_names, rotation=35, ha="right")
    ax.set_yticklabels(feature_names)

    for row in range(correlation.shape[0]):
        for column in range(correlation.shape[1]):
            ax.text(
                column,
                row,
                f"{correlation.iloc[row, column]:.2f}",
                ha="center",
                va="center",
            )

    ax.set_title("Feature Correlation Heatmap")
    fig.colorbar(image, ax=ax, label="Correlation")
    plt.tight_layout()
    plt.savefig(
        OUTPUT_DIR / "feature_correlation_heatmap.png",
        dpi=300,
        bbox_inches="tight",
    )
    plt.close()

    # Dataset summary
    summary_lines = [
        "IRIS DATASET SUMMARY",
        "=" * 60,
        f"Shape: {dataframe.shape}",
        "",
        "First five rows:",
        dataframe.head().to_string(index=False),
        "",
        "Class distribution:",
        dataframe["species"].value_counts().to_string(),
        "",
        "Descriptive statistics:",
        dataframe.describe().to_string(),
    ]
    save_text(OUTPUT_DIR / "dataset_summary.txt", "\n".join(summary_lines))


# ============================================================
# NUMPY MLP IMPLEMENTATION
# ============================================================

@dataclass
class TrainingHistory:
    """Stores values recorded during training."""

    epochs: List[int]
    training_loss: List[float]
    training_accuracy: List[float]
    validation_accuracy: List[float]


class NumPyMLP:
    """A fully connected multi-layer perceptron implemented with NumPy."""

    def __init__(
        self,
        layer_sizes: Sequence[int],
        hidden_activation: str = "sigmoid",
        learning_rate: float = 0.01,
        l2_lambda: float = 0.0,
        momentum: float = 0.0,
        random_seed: int = 42,
    ):
        if len(layer_sizes) < 3:
            raise ValueError("The network needs input, hidden, and output layers.")

        supported_activations = {"sigmoid", "tanh", "relu"}
        if hidden_activation not in supported_activations:
            raise ValueError(
                f"Activation must be one of {sorted(supported_activations)}."
            )

        self.layer_sizes = list(layer_sizes)
        self.hidden_activation = hidden_activation
        self.learning_rate = learning_rate
        self.l2_lambda = l2_lambda
        self.momentum = momentum
        self.random_generator = np.random.default_rng(random_seed)

        self.parameters = self.initialize_parameters()
        self.velocity = {
            key: np.zeros_like(value) for key, value in self.parameters.items()
        }

    def initialize_parameters(self) -> Dict[str, np.ndarray]:
        """
        Initialize weights with Xavier or He scaling and biases with zeros.
        """
        parameters: Dict[str, np.ndarray] = {}

        for layer_index in range(1, len(self.layer_sizes)):
            fan_in = self.layer_sizes[layer_index - 1]
            fan_out = self.layer_sizes[layer_index]

            if (
                self.hidden_activation == "relu"
                and layer_index < len(self.layer_sizes) - 1
            ):
                scale = math.sqrt(2.0 / fan_in)
            else:
                scale = math.sqrt(1.0 / fan_in)

            parameters[f"W{layer_index}"] = (
                self.random_generator.standard_normal((fan_in, fan_out)) * scale
            )
            parameters[f"b{layer_index}"] = np.zeros((1, fan_out))

        return parameters

    @staticmethod
    def sigmoid(values: np.ndarray) -> np.ndarray:
        """Numerically stable sigmoid activation."""
        clipped_values = np.clip(values, -500, 500)
        return 1.0 / (1.0 + np.exp(-clipped_values))

    @staticmethod
    def sigmoid_derivative(activated_values: np.ndarray) -> np.ndarray:
        """Derivative of sigmoid using its activated output."""
        return activated_values * (1.0 - activated_values)

    @staticmethod
    def tanh_derivative(activated_values: np.ndarray) -> np.ndarray:
        """Derivative of tanh using its activated output."""
        return 1.0 - np.square(activated_values)

    @staticmethod
    def relu(values: np.ndarray) -> np.ndarray:
        """ReLU activation."""
        return np.maximum(0.0, values)

    @staticmethod
    def relu_derivative(pre_activation_values: np.ndarray) -> np.ndarray:
        """Derivative of ReLU using pre-activation values."""
        return (pre_activation_values > 0).astype(float)

    @staticmethod
    def softmax(values: np.ndarray) -> np.ndarray:
        """Numerically stable softmax activation."""
        shifted_values = values - np.max(values, axis=1, keepdims=True)
        exponentials = np.exp(shifted_values)
        return exponentials / np.sum(exponentials, axis=1, keepdims=True)

    def activate_hidden(self, values: np.ndarray) -> np.ndarray:
        """Apply the selected hidden-layer activation."""
        if self.hidden_activation == "sigmoid":
            return self.sigmoid(values)
        if self.hidden_activation == "tanh":
            return np.tanh(values)
        return self.relu(values)

    def hidden_derivative(
        self,
        activated_values: np.ndarray,
        pre_activation_values: np.ndarray,
    ) -> np.ndarray:
        """Return the derivative for the selected hidden activation."""
        if self.hidden_activation == "sigmoid":
            return self.sigmoid_derivative(activated_values)
        if self.hidden_activation == "tanh":
            return self.tanh_derivative(activated_values)
        return self.relu_derivative(pre_activation_values)

    def forward_propagation(
        self,
        x: np.ndarray,
    ) -> Tuple[np.ndarray, Dict[str, np.ndarray]]:
        """Compute forward propagation through every layer."""
        cache: Dict[str, np.ndarray] = {"A0": x}
        number_of_weight_layers = len(self.layer_sizes) - 1

        for layer_index in range(1, number_of_weight_layers + 1):
            previous_activation = cache[f"A{layer_index - 1}"]
            weights = self.parameters[f"W{layer_index}"]
            bias = self.parameters[f"b{layer_index}"]

            z_values = previous_activation @ weights + bias
            cache[f"Z{layer_index}"] = z_values

            if layer_index == number_of_weight_layers:
                activation = self.softmax(z_values)
            else:
                activation = self.activate_hidden(z_values)

            cache[f"A{layer_index}"] = activation

        return cache[f"A{number_of_weight_layers}"], cache

    def compute_loss(
        self,
        predicted_probabilities: np.ndarray,
        actual_one_hot: np.ndarray,
    ) -> float:
        """Compute categorical cross-entropy loss with optional L2 penalty."""
        sample_count = actual_one_hot.shape[0]
        safe_probabilities = np.clip(predicted_probabilities, 1e-12, 1.0)

        cross_entropy = -np.sum(
            actual_one_hot * np.log(safe_probabilities)
        ) / sample_count

        l2_penalty = 0.0
        if self.l2_lambda > 0:
            for layer_index in range(1, len(self.layer_sizes)):
                l2_penalty += np.sum(
                    np.square(self.parameters[f"W{layer_index}"])
                )
            l2_penalty *= self.l2_lambda / (2.0 * sample_count)

        return float(cross_entropy + l2_penalty)

    def backward_propagation(
        self,
        actual_one_hot: np.ndarray,
        cache: Dict[str, np.ndarray],
    ) -> Dict[str, np.ndarray]:
        """Compute gradients using the chain rule."""
        gradients: Dict[str, np.ndarray] = {}
        sample_count = actual_one_hot.shape[0]
        final_layer = len(self.layer_sizes) - 1

        # Softmax combined with cross-entropy.
        dz = cache[f"A{final_layer}"] - actual_one_hot

        for layer_index in range(final_layer, 0, -1):
            previous_activation = cache[f"A{layer_index - 1}"]
            weights = self.parameters[f"W{layer_index}"]

            gradients[f"dW{layer_index}"] = (
                previous_activation.T @ dz
            ) / sample_count

            if self.l2_lambda > 0:
                gradients[f"dW{layer_index}"] += (
                    self.l2_lambda / sample_count
                ) * weights

            gradients[f"db{layer_index}"] = np.sum(
                dz,
                axis=0,
                keepdims=True,
            ) / sample_count

            if layer_index > 1:
                da_previous = dz @ weights.T
                previous_hidden_activation = cache[f"A{layer_index - 1}"]
                previous_hidden_z = cache[f"Z{layer_index - 1}"]

                dz = da_previous * self.hidden_derivative(
                    previous_hidden_activation,
                    previous_hidden_z,
                )

        return gradients

    def update_parameters(self, gradients: Dict[str, np.ndarray]) -> None:
        """Update parameters using gradient descent with optional momentum."""
        for layer_index in range(1, len(self.layer_sizes)):
            for parameter_name, gradient_name in (
                (f"W{layer_index}", f"dW{layer_index}"),
                (f"b{layer_index}", f"db{layer_index}"),
            ):
                if self.momentum > 0:
                    self.velocity[parameter_name] = (
                        self.momentum * self.velocity[parameter_name]
                        - self.learning_rate * gradients[gradient_name]
                    )
                    self.parameters[parameter_name] += self.velocity[parameter_name]
                else:
                    self.parameters[parameter_name] -= (
                        self.learning_rate * gradients[gradient_name]
                    )

    def predict_probabilities(self, x: np.ndarray) -> np.ndarray:
        """Return class probabilities."""
        probabilities, _ = self.forward_propagation(x)
        return probabilities

    def predict(self, x: np.ndarray) -> np.ndarray:
        """Return integer class predictions."""
        return np.argmax(self.predict_probabilities(x), axis=1)

    def score(self, x: np.ndarray, y: np.ndarray) -> float:
        """Return classification accuracy."""
        return float(np.mean(self.predict(x) == y))

    def fit(
        self,
        x_train: np.ndarray,
        y_train_one_hot: np.ndarray,
        y_train_labels: np.ndarray,
        x_validation: np.ndarray,
        y_validation_labels: np.ndarray,
        epochs: int = 1000,
        batch_size: int = 16,
        track_every: int = 50,
        early_stopping_patience: int = 8,
        minimum_improvement: float = 1e-4,
        verbose: bool = True,
    ) -> TrainingHistory:
        """
        Train using mini-batch gradient descent.

        Early stopping checks validation accuracy whenever metrics are recorded.
        The best parameter values are restored at the end.
        """
        history = TrainingHistory([], [], [], [])

        best_validation_accuracy = -np.inf
        best_parameters = {
            key: value.copy() for key, value in self.parameters.items()
        }
        checks_without_improvement = 0

        sample_count = x_train.shape[0]

        for epoch in range(1, epochs + 1):
            shuffled_indices = self.random_generator.permutation(sample_count)
            shuffled_x = x_train[shuffled_indices]
            shuffled_y = y_train_one_hot[shuffled_indices]

            for start_index in range(0, sample_count, batch_size):
                end_index = start_index + batch_size
                x_batch = shuffled_x[start_index:end_index]
                y_batch = shuffled_y[start_index:end_index]

                _, cache = self.forward_propagation(x_batch)
                gradients = self.backward_propagation(y_batch, cache)
                self.update_parameters(gradients)

            should_track = (
                epoch == 1
                or epoch % track_every == 0
                or epoch == epochs
            )

            if should_track:
                training_probabilities = self.predict_probabilities(x_train)
                training_loss = self.compute_loss(
                    training_probabilities,
                    y_train_one_hot,
                )
                training_accuracy = self.score(x_train, y_train_labels)
                validation_accuracy = self.score(
                    x_validation,
                    y_validation_labels,
                )

                history.epochs.append(epoch)
                history.training_loss.append(training_loss)
                history.training_accuracy.append(training_accuracy)
                history.validation_accuracy.append(validation_accuracy)

                if verbose:
                    print(
                        f"Epoch {epoch:4d} | "
                        f"Loss: {training_loss:.4f} | "
                        f"Train Accuracy: {training_accuracy:.4f} | "
                        f"Validation Accuracy: {validation_accuracy:.4f}"
                    )

                if (
                    validation_accuracy
                    > best_validation_accuracy + minimum_improvement
                ):
                    best_validation_accuracy = validation_accuracy
                    best_parameters = {
                        key: value.copy()
                        for key, value in self.parameters.items()
                    }
                    checks_without_improvement = 0
                else:
                    checks_without_improvement += 1

                if checks_without_improvement >= early_stopping_patience:
                    if verbose:
                        print(
                            "Early stopping activated at "
                            f"epoch {epoch}."
                        )
                    break

        self.parameters = best_parameters
        return history


# ============================================================
# VISUALIZATION AND EVALUATION
# ============================================================

def save_learning_curves(history: TrainingHistory) -> None:
    """Save the required training loss and accuracy curves."""

    plt.figure(figsize=(9, 6))
    plt.plot(
        history.epochs,
        history.training_loss,
        marker="o",
        label="Training Loss",
    )
    plt.xlabel("Epoch")
    plt.ylabel("Cross-Entropy Loss")
    plt.title("Training Loss Curve")
    plt.grid(alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.savefig(
        OUTPUT_DIR / "training_loss_curve.png",
        dpi=300,
        bbox_inches="tight",
    )
    plt.close()

    plt.figure(figsize=(9, 6))
    plt.plot(
        history.epochs,
        history.training_accuracy,
        marker="o",
        label="Training Accuracy",
    )
    plt.plot(
        history.epochs,
        history.validation_accuracy,
        marker="s",
        label="Validation Accuracy",
    )
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.ylim(0.0, 1.05)
    plt.title("Training and Validation Accuracy Curves")
    plt.grid(alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.savefig(
        OUTPUT_DIR / "training_validation_accuracy_curves.png",
        dpi=300,
        bbox_inches="tight",
    )
    plt.close()


def save_confusion_matrix_plot(
    actual_labels: np.ndarray,
    predicted_labels: np.ndarray,
    class_names: Sequence[str],
) -> np.ndarray:
    """Save a confusion matrix heatmap and return the matrix."""
    matrix = confusion_matrix(actual_labels, predicted_labels)

    fig, ax = plt.subplots(figsize=(7, 6))
    image = ax.imshow(matrix, cmap="Blues")

    ax.set_xticks(range(len(class_names)))
    ax.set_yticks(range(len(class_names)))
    ax.set_xticklabels(class_names, rotation=25, ha="right")
    ax.set_yticklabels(class_names)
    ax.set_xlabel("Predicted Label")
    ax.set_ylabel("True Label")
    ax.set_title("NumPy MLP Confusion Matrix")

    for row in range(matrix.shape[0]):
        for column in range(matrix.shape[1]):
            ax.text(
                column,
                row,
                str(matrix[row, column]),
                ha="center",
                va="center",
                fontsize=12,
            )

    fig.colorbar(image, ax=ax)
    plt.tight_layout()
    plt.savefig(
        OUTPUT_DIR / "confusion_matrix_heatmap.png",
        dpi=300,
        bbox_inches="tight",
    )
    plt.close()

    return matrix


def save_classification_outputs(
    actual_labels: np.ndarray,
    predicted_labels: np.ndarray,
    class_names: Sequence[str],
) -> pd.DataFrame:
    """Save classification report as CSV and text."""
    report_dictionary = classification_report(
        actual_labels,
        predicted_labels,
        target_names=class_names,
        output_dict=True,
        zero_division=0,
    )
    report_dataframe = pd.DataFrame(report_dictionary).transpose()
    report_dataframe.to_csv(
        OUTPUT_DIR / "classification_report.csv",
        index=True,
    )

    report_text = classification_report(
        actual_labels,
        predicted_labels,
        target_names=class_names,
        zero_division=0,
    )
    save_text(OUTPUT_DIR / "classification_report.txt", report_text)

    return report_dataframe


def create_pca_decision_boundary(
    x_all_scaled: np.ndarray,
    y_all: np.ndarray,
    class_names: Sequence[str],
) -> None:
    """
    Project standardized features into two PCA dimensions, train a separate
    NumPy MLP in that 2D space, and save its decision boundary.
    """
    pca = PCA(n_components=2, random_state=RANDOM_SEED)
    x_pca = pca.fit_transform(x_all_scaled)

    x_train, x_test, y_train, y_test = train_test_split(
        x_pca,
        y_all,
        test_size=0.20,
        random_state=RANDOM_SEED,
        stratify=y_all,
    )

    y_train_one_hot = one_hot_encode(y_train, len(class_names))

    boundary_model = NumPyMLP(
        layer_sizes=[2, 8, 6, 3],
        hidden_activation="tanh",
        learning_rate=0.01,
        l2_lambda=0.0001,
        momentum=0.9,
        random_seed=RANDOM_SEED,
    )

    boundary_model.fit(
        x_train=x_train,
        y_train_one_hot=y_train_one_hot,
        y_train_labels=y_train,
        x_validation=x_test,
        y_validation_labels=y_test,
        epochs=1000,
        batch_size=16,
        track_every=50,
        early_stopping_patience=10,
        verbose=False,
    )

    padding = 0.8
    x_min, x_max = x_pca[:, 0].min() - padding, x_pca[:, 0].max() + padding
    y_min, y_max = x_pca[:, 1].min() - padding, x_pca[:, 1].max() + padding

    grid_x, grid_y = np.meshgrid(
        np.linspace(x_min, x_max, 400),
        np.linspace(y_min, y_max, 400),
    )
    grid_points = np.column_stack((grid_x.ravel(), grid_y.ravel()))
    grid_predictions = boundary_model.predict(grid_points).reshape(grid_x.shape)

    plt.figure(figsize=(10, 7))
    plt.contourf(
        grid_x,
        grid_y,
        grid_predictions,
        alpha=0.25,
        levels=np.arange(len(class_names) + 1) - 0.5,
        cmap="viridis",
    )

    for class_index, class_name in enumerate(class_names):
        class_mask = y_all == class_index
        plt.scatter(
            x_pca[class_mask, 0],
            x_pca[class_mask, 1],
            label=class_name,
            edgecolors="black",
            s=45,
        )

    explained_variance = pca.explained_variance_ratio_ * 100
    plt.xlabel(f"Principal Component 1 ({explained_variance[0]:.1f}% variance)")
    plt.ylabel(f"Principal Component 2 ({explained_variance[1]:.1f}% variance)")
    plt.title("PCA Decision Boundary of NumPy MLP")
    plt.legend()
    plt.tight_layout()
    plt.savefig(
        OUTPUT_DIR / "pca_decision_boundary.png",
        dpi=300,
        bbox_inches="tight",
    )
    plt.close()


# ============================================================
# EXPERIMENTS
# ============================================================

def run_single_experiment(
    name: str,
    architecture: Sequence[int],
    activation: str,
    learning_rate: float,
    x_train: np.ndarray,
    y_train: np.ndarray,
    y_train_one_hot: np.ndarray,
    x_validation: np.ndarray,
    y_validation: np.ndarray,
    x_test: np.ndarray,
    y_test: np.ndarray,
) -> Dict[str, object]:
    """Train and evaluate one experimental configuration."""
    model = NumPyMLP(
        layer_sizes=architecture,
        hidden_activation=activation,
        learning_rate=learning_rate,
        l2_lambda=0.0001,
        momentum=0.0,
        random_seed=RANDOM_SEED,
    )

    history = model.fit(
        x_train=x_train,
        y_train_one_hot=y_train_one_hot,
        y_train_labels=y_train,
        x_validation=x_validation,
        y_validation_labels=y_validation,
        epochs=BASELINE_EPOCHS,
        batch_size=BATCH_SIZE,
        track_every=TRACK_EVERY,
        early_stopping_patience=8,
        verbose=False,
    )

    predictions = model.predict(x_test)
    test_accuracy = accuracy_score(y_test, predictions)
    weighted_precision, weighted_recall, weighted_f1, _ = (
        precision_recall_fscore_support(
            y_test,
            predictions,
            average="weighted",
            zero_division=0,
        )
    )

    return {
        "configuration": name,
        "architecture": "-".join(str(size) for size in architecture),
        "activation": activation,
        "learning_rate": learning_rate,
        "epochs_completed": history.epochs[-1],
        "final_training_loss": history.training_loss[-1],
        "best_validation_accuracy": max(history.validation_accuracy),
        "test_accuracy": test_accuracy,
        "weighted_precision": weighted_precision,
        "weighted_recall": weighted_recall,
        "weighted_f1_score": weighted_f1,
    }


def run_experiments(data: Dict[str, object]) -> pd.DataFrame:
    """Run the assignment's learning-rate, activation, and architecture tests."""
    configurations = [
        # Learning-rate experiments
        ("LR 0.001", [4, 8, 6, 3], "sigmoid", 0.001),
        ("LR 0.01", [4, 8, 6, 3], "sigmoid", 0.01),
        ("LR 0.1", [4, 8, 6, 3], "sigmoid", 0.1),

        # Activation experiments
        ("Activation Sigmoid", [4, 8, 6, 3], "sigmoid", 0.01),
        ("Activation Tanh", [4, 8, 6, 3], "tanh", 0.01),
        ("Activation ReLU", [4, 8, 6, 3], "relu", 0.01),

        # Architecture experiments
        ("Shallow 4-8-3", [4, 8, 3], "tanh", 0.01),
        ("Required 4-8-6-3", [4, 8, 6, 3], "tanh", 0.01),
        ("Deeper 4-12-8-6-3", [4, 12, 8, 6, 3], "tanh", 0.01),
    ]

    rows = []

    print("\nRunning required experiments...")
    for index, (name, architecture, activation, learning_rate) in enumerate(
        configurations,
        start=1,
    ):
        print(f"  Experiment {index}/{len(configurations)}: {name}")
        row = run_single_experiment(
            name=name,
            architecture=architecture,
            activation=activation,
            learning_rate=learning_rate,
            x_train=data["x_train"],
            y_train=data["y_train"],
            y_train_one_hot=data["y_train_one_hot"],
            x_validation=data["x_validation"],
            y_validation=data["y_validation"],
            x_test=data["x_test"],
            y_test=data["y_test"],
        )
        rows.append(row)

    results = pd.DataFrame(rows)
    results = results.sort_values(
        by="test_accuracy",
        ascending=False,
    ).reset_index(drop=True)

    results.to_csv(
        OUTPUT_DIR / "experiment_comparison_table.csv",
        index=False,
    )

    # Visual comparison of test accuracy
    plot_data = results.sort_values(by="test_accuracy", ascending=True)

    plt.figure(figsize=(11, 7))
    bars = plt.barh(
        plot_data["configuration"],
        plot_data["test_accuracy"],
    )
    plt.xlim(0.0, 1.05)
    plt.xlabel("Test Accuracy")
    plt.title("Experimental Configuration Comparison")

    for bar, value in zip(bars, plot_data["test_accuracy"]):
        plt.text(
            value + 0.01,
            bar.get_y() + bar.get_height() / 2,
            f"{value:.3f}",
            va="center",
        )

    plt.tight_layout()
    plt.savefig(
        OUTPUT_DIR / "experiment_comparison_graph.png",
        dpi=300,
        bbox_inches="tight",
    )
    plt.close()

    return results


# ============================================================
# SKLEARN COMPARISON
# ============================================================

def compare_with_sklearn(
    baseline_model: NumPyMLP,
    data: Dict[str, object],
) -> pd.DataFrame:
    """Compare the custom NumPy MLP with sklearn's MLPClassifier."""
    numpy_predictions = baseline_model.predict(data["x_test"])

    sklearn_model = MLPClassifier(
        hidden_layer_sizes=(8, 6),
        activation="logistic",
        solver="sgd",
        learning_rate_init=BASELINE_LEARNING_RATE,
        batch_size=BATCH_SIZE,
        max_iter=3000,
        random_state=RANDOM_SEED,
        early_stopping=True,
        validation_fraction=0.15,
        n_iter_no_change=50,
    )
    sklearn_model.fit(data["x_train"], data["y_train"])
    sklearn_predictions = sklearn_model.predict(data["x_test"])

    rows = []

    for model_name, predictions in [
        ("NumPy MLP from scratch", numpy_predictions),
        ("sklearn MLPClassifier", sklearn_predictions),
    ]:
        precision, recall, f1_score, _ = precision_recall_fscore_support(
            data["y_test"],
            predictions,
            average="weighted",
            zero_division=0,
        )

        rows.append(
            {
                "model": model_name,
                "accuracy": accuracy_score(data["y_test"], predictions),
                "weighted_precision": precision,
                "weighted_recall": recall,
                "weighted_f1_score": f1_score,
            }
        )

    comparison = pd.DataFrame(rows)
    comparison.to_csv(
        OUTPUT_DIR / "numpy_vs_sklearn_comparison.csv",
        index=False,
    )

    metric_columns = [
        "accuracy",
        "weighted_precision",
        "weighted_recall",
        "weighted_f1_score",
    ]

    x_positions = np.arange(len(metric_columns))
    bar_width = 0.35

    plt.figure(figsize=(10, 6))
    plt.bar(
        x_positions - bar_width / 2,
        comparison.loc[0, metric_columns],
        width=bar_width,
        label=comparison.loc[0, "model"],
    )
    plt.bar(
        x_positions + bar_width / 2,
        comparison.loc[1, metric_columns],
        width=bar_width,
        label=comparison.loc[1, "model"],
    )
    plt.xticks(
        x_positions,
        ["Accuracy", "Precision", "Recall", "F1-score"],
    )
    plt.ylim(0.0, 1.05)
    plt.ylabel("Score")
    plt.title("NumPy MLP vs sklearn MLPClassifier")
    plt.legend()
    plt.tight_layout()
    plt.savefig(
        OUTPUT_DIR / "numpy_vs_sklearn_comparison.png",
        dpi=300,
        bbox_inches="tight",
    )
    plt.close()

    return comparison


# ============================================================
# WRITTEN REPORT
# ============================================================

def create_written_report(
    baseline_accuracy: float,
    experiment_results: pd.DataFrame,
    sklearn_comparison: pd.DataFrame,
    class_names: Sequence[str],
) -> None:
    """Create a concise assignment report in Markdown format."""
    best_row = experiment_results.iloc[0]
    sklearn_accuracy = float(
        sklearn_comparison.loc[
            sklearn_comparison["model"] == "sklearn MLPClassifier",
            "accuracy",
        ].iloc[0]
    )

    learning_rate_rows = experiment_results[
        experiment_results["configuration"].str.startswith("LR ")
    ].sort_values("learning_rate")

    activation_rows = experiment_results[
        experiment_results["configuration"].str.startswith("Activation ")
    ].sort_values("test_accuracy", ascending=False)

    architecture_rows = experiment_results[
        experiment_results["configuration"].isin(
            [
                "Shallow 4-8-3",
                "Required 4-8-6-3",
                "Deeper 4-12-8-6-3",
            ]
        )
    ].sort_values("test_accuracy", ascending=False)

    best_activation = activation_rows.iloc[0]["activation"]
    best_architecture = architecture_rows.iloc[0]["architecture"]

    learning_rate_summary = "\n".join(
        (
            f"- Learning rate {row.learning_rate}: "
            f"test accuracy {row.test_accuracy:.4f}, "
            f"final loss {row.final_training_loss:.4f}."
        )
        for row in learning_rate_rows.itertuples()
    )

    architecture_summary = "\n".join(
        (
            f"- {row.configuration}: test accuracy "
            f"{row.test_accuracy:.4f}."
        )
        for row in architecture_rows.itertuples()
    )

    report = f"""# Assignment 1 Report: ANN Fundamentals Using NumPy

## 1. Objective

The objective was to implement a multi-layer perceptron from scratch using
NumPy and classify the three Iris species: {", ".join(class_names)}.
The implementation includes parameter initialization, forward propagation,
cross-entropy loss, backpropagation, mini-batch gradient descent, prediction,
evaluation, and early stopping.

## 2. Data Preparation

The Iris dataset contains 150 observations and four numerical input features.
A stratified split was used to create 70% training data, 15% validation data,
and 15% test data. Z-score normalization was fitted only on the training set
and then applied to validation and test data. Labels were converted into
three-element one-hot vectors.

The generated exploratory outputs include feature histograms, a pair plot,
and a feature-correlation heatmap. Petal length and petal width show a strong
positive relationship and provide substantial class-separation information.

## 3. Network Architecture

The required baseline architecture was used:

- Input layer: 4 neurons
- Hidden layer 1: 8 sigmoid neurons
- Hidden layer 2: 6 sigmoid neurons
- Output layer: 3 softmax neurons

Weights were initialized using scaled random values. Hidden layers used the
selected activation function, while the output layer used softmax to obtain
class probabilities.

## 4. Forward Propagation and Loss

For each layer, the linear output was calculated as:

`Z = A_previous × W + b`

The hidden activation was then applied to produce the next activation.
The final layer used softmax. Categorical cross-entropy measured the error
between predicted probabilities and one-hot encoded labels.

## 5. Backpropagation

The output-layer gradient simplifies to predicted probabilities minus the
one-hot target when softmax is combined with cross-entropy. Gradients were
propagated backward through the network using the chain rule. Weight gradients
were calculated from the previous layer's activations and the current layer's
error term. Parameters were updated with mini-batch gradient descent.

## 6. Baseline Results

The custom NumPy network achieved a test accuracy of
**{baseline_accuracy:.4f}**. Precision, recall, and F1-score for each class are
available in `classification_report.csv` and `classification_report.txt`.
The confusion-matrix heatmap shows the distribution of correct and incorrect
predictions.

## 7. Hyperparameter Experiments

### Learning Rate

{learning_rate_summary}

A very small learning rate generally changes parameters slowly and may need
more epochs. A larger learning rate converges faster but can become unstable
if updates overshoot useful parameter values.

### Activation Function

The strongest activation in this run was **{best_activation}**. Sigmoid is
smooth but may produce small gradients in deeper networks. Tanh is zero-centered
and often trains faster on standardized data. ReLU avoids saturation for
positive inputs but can produce inactive neurons when many pre-activations are
negative.

### Network Depth

{architecture_summary}

The strongest tested architecture was **{best_architecture}**. Additional
depth increases model capacity, but Iris is a small dataset, so an unnecessarily
deep network can overfit or become harder to optimize.

### Batch Size

A mini-batch size of 16 was used. It provides more stable gradients than
single-sample updates while still introducing enough variation to help
optimization. Very large batches would make updates smoother but reduce the
number of updates per epoch.

## 8. Comparison with sklearn

The NumPy model achieved {baseline_accuracy:.4f} test accuracy, while sklearn's
MLPClassifier achieved {sklearn_accuracy:.4f}. Small differences are expected
because sklearn includes its own convergence rules and optimized numerical
implementation. The comparison confirms whether the manually implemented
forward and backward calculations behave reasonably.

## 9. Best Configuration

The highest test accuracy in the experiment table was
**{float(best_row["test_accuracy"]):.4f}**, produced by:

- Configuration: {best_row["configuration"]}
- Architecture: {best_row["architecture"]}
- Activation: {best_row["activation"]}
- Learning rate: {best_row["learning_rate"]}

## 10. Conclusion

The assignment demonstrates that a complete neural network can be built using
basic NumPy matrix operations. Standardization, suitable initialization,
activation choice, learning rate, and network depth all influence convergence.
For Iris, a modest architecture is sufficient because the dataset is small and
the classes are reasonably separable, especially through petal measurements.
"""

    save_text(OUTPUT_DIR / "written_report.md", report)


# ============================================================
# MAIN PROGRAM
# ============================================================

def main() -> None:
    """Run the entire assignment and save only required deliverables."""
    create_output_folder()

    print("=" * 70)
    print("IRIS CLASSIFICATION USING A NUMPY MLP")
    print("=" * 70)
    print(f"Project folder: {PROJECT_DIR}")
    print(f"Output folder : {OUTPUT_DIR}")

    data = load_and_prepare_data()

    print("\nDataset loaded successfully.")
    print(f"Training samples  : {data['x_train'].shape[0]}")
    print(f"Validation samples: {data['x_validation'].shape[0]}")
    print(f"Test samples      : {data['x_test'].shape[0]}")

    save_eda_outputs(
        dataframe=data["dataframe"],
        feature_names=data["feature_names"],
    )

    print("\nTraining required baseline network...")
    baseline_model = NumPyMLP(
        layer_sizes=[4, 8, 6, 3],
        hidden_activation="sigmoid",
        learning_rate=BASELINE_LEARNING_RATE,
        l2_lambda=0.0,
        momentum=0.0,
        random_seed=RANDOM_SEED,
    )

    baseline_history = baseline_model.fit(
        x_train=data["x_train"],
        y_train_one_hot=data["y_train_one_hot"],
        y_train_labels=data["y_train"],
        x_validation=data["x_validation"],
        y_validation_labels=data["y_validation"],
        epochs=BASELINE_EPOCHS,
        batch_size=BATCH_SIZE,
        track_every=TRACK_EVERY,
        early_stopping_patience=8,
        verbose=True,
    )

    save_learning_curves(baseline_history)

    test_predictions = baseline_model.predict(data["x_test"])
    test_accuracy = accuracy_score(data["y_test"], test_predictions)

    confusion = save_confusion_matrix_plot(
        actual_labels=data["y_test"],
        predicted_labels=test_predictions,
        class_names=data["class_names"],
    )

    save_classification_outputs(
        actual_labels=data["y_test"],
        predicted_labels=test_predictions,
        class_names=data["class_names"],
    )

    # Save sample-level test predictions as part of evaluation evidence.
    prediction_table = pd.DataFrame(
        {
            "actual_class": [
                data["class_names"][index] for index in data["y_test"]
            ],
            "predicted_class": [
                data["class_names"][index] for index in test_predictions
            ],
            "correct": data["y_test"] == test_predictions,
        }
    )
    prediction_table.to_csv(
        OUTPUT_DIR / "test_predictions.csv",
        index=False,
    )

    # Standardize all rows with the already-fitted training scaler for PCA plot.
    x_all_scaled = data["scaler"].transform(data["x"])
    create_pca_decision_boundary(
        x_all_scaled=x_all_scaled,
        y_all=data["y"],
        class_names=data["class_names"],
    )

    experiment_results = run_experiments(data)
    sklearn_comparison = compare_with_sklearn(baseline_model, data)

    create_written_report(
        baseline_accuracy=test_accuracy,
        experiment_results=experiment_results,
        sklearn_comparison=sklearn_comparison,
        class_names=data["class_names"],
    )

    metrics_summary = {
        "baseline_test_accuracy": float(test_accuracy),
        "training_samples": int(data["x_train"].shape[0]),
        "validation_samples": int(data["x_validation"].shape[0]),
        "test_samples": int(data["x_test"].shape[0]),
        "baseline_architecture": [4, 8, 6, 3],
        "baseline_activation": "sigmoid",
        "baseline_learning_rate": BASELINE_LEARNING_RATE,
        "batch_size": BATCH_SIZE,
        "maximum_epochs": BASELINE_EPOCHS,
        "epochs_completed": int(baseline_history.epochs[-1]),
        "confusion_matrix": confusion.tolist(),
    }

    # A plain text summary is required; JSON is intentionally not produced.
    summary_text = "\n".join(
        [
            "FINAL RESULTS",
            "=" * 60,
            f"Baseline test accuracy: {test_accuracy:.4f}",
            f"Architecture: {metrics_summary['baseline_architecture']}",
            f"Activation: {metrics_summary['baseline_activation']}",
            f"Learning rate: {BASELINE_LEARNING_RATE}",
            f"Batch size: {BATCH_SIZE}",
            f"Epochs completed: {baseline_history.epochs[-1]}",
            "",
            "Confusion matrix:",
            np.array2string(confusion),
            "",
            "Best experiment:",
            experiment_results.iloc[0].to_string(),
        ]
    )
    save_text(OUTPUT_DIR / "final_results.txt", summary_text)

    required_files = [
        "dataset_summary.txt",
        "feature_histograms.png",
        "feature_pair_plot.png",
        "feature_correlation_heatmap.png",
        "training_loss_curve.png",
        "training_validation_accuracy_curves.png",
        "confusion_matrix_heatmap.png",
        "classification_report.csv",
        "classification_report.txt",
        "test_predictions.csv",
        "pca_decision_boundary.png",
        "experiment_comparison_table.csv",
        "experiment_comparison_graph.png",
        "numpy_vs_sklearn_comparison.csv",
        "numpy_vs_sklearn_comparison.png",
        "written_report.md",
        "final_results.txt",
    ]

    print("\n" + "=" * 70)
    print(f"Baseline test accuracy: {test_accuracy:.4f}")
    print("Required outputs created:")
    for filename in required_files:
        print(f"  - {filename}")
    print(f"\nAll files were saved in:\n{OUTPUT_DIR}")
    print("=" * 70)


if __name__ == "__main__":
    main()
```
